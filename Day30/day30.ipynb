# Importing dataset from seaborn library

```python
df = sns.load_dataset('tips')
df.head()
```

### Barplot()

```python
sns.barplot(x = 'day', y = 'total_bill', data=df)
```

```python
# Horizontal bar plot
sns.barplot(x = 'total_bill',y = 'day', data=df)
```

#### Hue() attribute
- The hue attribute splits the bars based on different values in a particular column


```python
sns.barplot(x='day',y='total_bill', hue='sex', data=df)
```

#### Palette attribute
- deep
- winter_r
- spring

```python
# deep
sns.palplot(sns.color_palette('deep',10))
```

```python
# spring
sns.palplot(sns.color_palette('spring',10))
```

```python
# winter_r
sns.palplot(sns.color_palette('winter_r',10))
```

#### Creating custom palettes

```python
custom = ['red','blue','green','pink']
sns.set_palette(custom)
sns.palplot(sns.color_palette())
```

#### Order() attribute


```python
sns.barplot(x='day',y='total_bill', data=df)
```

```python
sns.barplot(x='day',y='total_bill', order = ['Sun', 'Thur', 'Fri', 'Sat'], data=df)
```

### Estimator attribute

```python
sns.barplot(x='day',y='total_bill', estimator=np.median, data=df)
```

### Dist plots
- We can use displot we get KDE, histogram & rug plot in a single plot itself

```python
df_penguins = sns.load_dataset('penguins')
```

```python
df_penguins.head(10)
```

```python
## We need to use rug = True to get rug plot
sns.distplot(df_penguins['body_mass_g'], rug=True)
```

```python
# Removing KDE plot
sns.distplot(df_penguins['body_mass_g'], color = 'black', kde=False)
```

```python
sns.distplot(df_penguins["body_mass_g"],color = 'black' ,hist = False)
```

```python
sns.distplot(df_penguins["body_mass_g"], rug = True, bins = [3200,3500,3900,5000,5500,5800,6000])
```

Styling rugplots

```python
sns.distplot(df_penguins['body_mass_g'], rug=True, rug_kws ={'color':'red','height':0.3}, hist = False, kde = False)
```

Styling histograms

```python
sns.distplot(df_penguins['body_mass_g'], hist_kws = {'color':'red', 'histtype':'step'}, kde=False)
```

Styling KDE plot

```python
sns.distplot(df_penguins["body_mass_g"],kde_kws = {"color":"green","alpha":1},hist = False)
```

Styling KDE, hist and rug in a single plot

```python
sns.distplot(df_penguins["bill_length_mm"],rug=True, rug_kws={"color":"red","height":0.2},kde_kws={"color":"black","alpha":1}, hist_kws={"color":"blue"})
```

Color Attribute

```python
# Horizontal boxplots and ordering of bars
sns.boxplot(y = df_penguins["island"], x = df_penguins["flipper_length_mm"],hue=df_penguins['sex'], order = ["Torgersen","Biscoe","Dream"], color='g', width=0.8)
```

whis

- By default whis value is 0.8

```python
# Horizontal boxplots and ordering of bars
sns.boxplot(y = df_penguins["island"], x = df_penguins["flipper_length_mm"],hue=df_penguins['sex'], order = ["Torgersen","Biscoe","Dream"], color='pink', whis = 0.3)
```

Showcaps

- showcaps by default takes boolean values

```python
# Horizontal boxplots and ordering of bars
sns.boxplot(y = df_penguins["island"], x = df_penguins["flipper_length_mm"],hue=df_penguins['sex'], order = ["Torgersen","Biscoe","Dream"], color='pink', showcaps = True)
```
