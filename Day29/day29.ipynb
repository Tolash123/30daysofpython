### Axes labels, Legend, Grid

```python
days = [1,2,3,4,5,6,7]         # Days of the week
max_t = [50,51,52,53,47,49,46] # Maximum Temperature
min_t = [43,42,40,35,34,37,32] # Minimum Temperature
avg_t = [45,48,46,40,42,41,45] # Averagee Temperature
```

```python
# plotting days on x-axis and other 3 lists on y-axis
plt.xlabel('Days')
plt.ylabel('Temperatures')
plt.title('Weather')
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(avg_t)
```

- From above graph, we don't know which is max_t, min_t. We need legend
Legend()

```python
plt.xlabel("Days")
plt.ylabel("Temperatures")
plt.title("Weather")
plt.plot(days,max_t ,label="max_t")
plt.plot(days,min_t ,label="min_t")
plt.plot(avg_t ,label = "avg_t")

plt.legend()
```

From above graph we got legend at top right side.
We can change dat as well

```python
plt.xlabel("Days")
plt.ylabel("Temperatures")
plt.title("Weather")
plt.plot(days,max_t ,label="max_t")
plt.plot(days,min_t ,label="min_t")
plt.plot(avg_t ,label = "avg_t")

plt.legend(loc='upper left', shadow=True, fontsize='small')
# Note: you can always change the position of the legend from right to left
```

#### Grid()
Adding grid to our graph

```python
plt.xlabel("Days")
plt.ylabel("Temperatures")
plt.title("Weather")
plt.plot(days,max_t ,label="max_t")
plt.plot(days,min_t ,label="min_t")
plt.plot(avg_t ,label = "avg_t")

plt.legend()
plt.grid()
```

#### Bar plot

```python
company = ['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue = [90,136,89,27]
profit= [40,2,34,12]
```

```python
# since elements in company is strings, we convert to ndarrays
xc = np.arange(len(company))
xc
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company's Revenue")
plt.xticks(xc, company) # To get company names on x-axis
plt.bar(company,revenue, color='g', label='Revenue')
plt.legend()
```

#### Plotting 2 different values in same plot

```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company's Revenue")
plt.xticks(xc, company) # To get company names on x-axis
plt.bar(company,revenue, color='b', label='Revenue')
plt.bar(company,profit, color='r', label='profit')
plt.legend()
```

#### To get graphs side by side in single plot


```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company's Revenue")
plt.xticks(xc, company) # To get company names on x-axis
plt.bar(xc-0.3,revenue, width=0.4, color='b', label='Revenue')
plt.bar(xc+0.1, profit, width=0.4, color='r', label='profit')
plt.legend()
```

#### Horizontal Bar Chart
- barh

```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Company's Revenue")
plt.yticks(xc, company) # To get company names on x-axis
plt.barh(xc-0.3,revenue, color='b', label='Revenue')
plt.barh(xc+0.1, profit, color='r', label='profit')
plt.legend()
plt.show()
```

#### Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=4,rwidth=0.5) # By default y axis is frequency
plt.show()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=[80,100,125,200],rwidth=0.9) # bins range
plt.show()
```

#### Orientation, plots histogram horizontally


```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=[80,100,125,200],color = 'pink',rwidth=0.5, orientation='horizontal') # bins range
plt.show()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women],label = ['Men','Women'])
plt.legend()
plt.show()
```

#### Pie-Charts

```python
value = [1400, 600, 300, 410, 250]
value_labels = ['Home Rent', 'Food', 'Internet Bill', 'Car', 'Electricity']
plt.pie(value,labels=value_labels, radius=2)
plt.show()
```

To show '%' in pie charts
autopct = '%.2f%%'

```python
value = [1400, 600, 300, 410, 250]
value_labels = ['Home Rent', 'Food', 'Internet Bill', 'Car', 'Electricity']
plt.pie(value,labels=value_labels, radius=2, autopct='%.2f%%')
plt.show()
```

```python
# Explode
value = [1400, 600, 300, 410, 250]
value_labels = ['Home Rent', 'Food', 'Internet Bill', 'Car', 'Electricity']
plt.pie(value,labels=value_labels, radius=2, autopct='%.2f%%', explode=[0,0.3,0,0,0])
plt.show()
```

##### Rotating Pie Chart

```python
# Startangle
value = [1400, 600, 300, 410, 250]
value_labels = ['Home Rent', 'Food', 'Internet Bill', 'Car', 'Electricity']
plt.pie(value,labels=value_labels, radius=2, autopct='%.2f%%', explode=[0,0.3,0,0,0], startangle=180)
plt.show()
```
