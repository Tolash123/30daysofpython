```python
# For us to be able to group the data and do some analysis
# we can do that by deriving month from the Date
df_aapl = pd.read_csv('aapl.csv')
df_aapl['Date'] = pd.to_datetime(df_aapl['Date'])
df_aapl['Month']= df_aapl['Date'].dt.month_name()
df_aapl.head()
```

```python
# we want our output not to the the month_name instead number 
df_aapl = pd.read_csv('aapl.csv')
df_aapl['Date'] = pd.to_datetime(df_aapl['Date'])
df_aapl['Month']= df_aapl['Date'].dt.month
df_aapl.head()
```

```python
df_aapl.groupby(by='Month').Open.mean()
# This is to know the average open price for each month
```

```python
df_aapl.groupby(by='Month').Open.std()
# To compute the standard deviation of each months
```

```python
# To know both the average open price and the average volume in each month
df_aapl.groupby(by='Month')[['Open','Volume']].mean()

# Note: if we use df_aapl.groupby(by='Month')['Open','Volume'].mean()
# It won't work, it has to be in list
```

```python
# Let's check for the average open price for each month and each day of the week
df_aapl['Weekday'] = df_aapl['Date'].dt.day_name()
df_aapl.head()
# Note: .dt.weekday_name not working instead we use dt.day_name()
```

```python
# To know the average open price for each day and month
df_aapl.groupby(by=['Month','Weekday']).Open.mean()
```

```python
# To know both the average open price and the average volume in each month and week days
df_aapl.groupby(by=['Month','Weekday'])[['Open','Volume']].mean()
```

We now have a group for every month + weekday combination. The above is a series with multi-level

```python
# To reset back the index
df_aapl.groupby(by=['Month', 'Weekday'])[['Open','Volume']].mean().reset_index()
```

### Aggregate

we can use the aggregate method to summarize the groups using multiple functions. let's assume we want to know the avg and standard deviation of the open price for each month. After the groupby, we use .agg and then we can specify the list of functions that will be used to summarize each group.

```python
df_aapl.groupby(by='Month').Open.agg( [np.mean, np.std])
```

```python
output = df_aapl.groupby(by='Month')[['Open','Volume']].agg([np.mean, np.std])
output
```

```python
# To show a dataframe with multi-level colums which can be sliced
output.loc[:, 'Volume']
```

```python
# Using the dictionary
df_aapl.groupby(by='Month').agg({'Open': np.mean, 'Volume': np.max})
```

```python
# We can use the agg() with  customer function. let's count the number of days in each month, if the open price was above the average open price for the month  
# Using customer function
def compare_open_price(col):
    count =0
    avg_price = col.mean()
    for i in list(col.index):
        open_price = col[i]
        if open_price>=avg_price:
            count=count+1
    return count
df_aapl.groupby(by='Month').agg({'Open': compare_open_price})
```

### Transform 
Instead of aggregating each group we can apply a transformation with transform() method after the groupby. The transform() method takes a column and returns back a series that is the same length

```python
# Using the transform() method with the sum function on the column volume
volume_by_month = df_aapl.groupby(by='Month').Volume.transform(sum)
volume_by_month
```

```python
volume_by_month= df_aapl.groupby(by='Month').Volume.agg(sum)
volume_by_month
```

### Filtering

The filter() method after a groupby let's us only select rows corresponding to each group where a certain criterion regarding the group as a whole is true is satisfied. The filter() method must take in a function returns a boolean. The function will be run once for each row.

```python
df_aapl.groupby(by='Month').agg({'Open': np.mean})
```

```python
df_aapl.groupby(by='Month').filter(lambda x: x['Open'].mean()>=10)
```
